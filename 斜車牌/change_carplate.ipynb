```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
img_plate = cv2.imread('ROI-009.jpg')
plt.imshow(img_plate)
plt.show()
rows, cols = img_plate.shape[:2]

# original pts
pts_o = np.float32([[6, 8], [95, 39], [5, 38], [94, 73]]) # 这四个点为原始图片上数独的位置
pts_d = np.float32([[0, 0], [90, 0], [0, 30], [90, 30]]) # 这是变换之后的图上四个点的位置

# get transform matrix
M = cv2.getPerspectiveTransform(pts_o, pts_d)
# apply transformation
correct_plate = cv2.warpPerspective(img_plate, M, (90,30)) # 最后一参数是输出dst的尺寸。可以和原来图片尺寸不一致。按需求来确定
#dst = cv2.resize(dst, (600,200))
plt.imshow( correct_plate)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

```python
car = cv2.imread('22.png')
print(car.shape)
plt.imshow(car)
```

```python
tmp_plate = cv2.resize(correct_plate, (922,624))
pt_o = np.float32([[0, 0], [922, 0], [0, 624], [922, 624]]) # 这四个点为原始图片上数独的位置
pt_d = np.float32([[130, 382], [231, 407], [125, 435], [227, 460]]) # 这是变换之后的图上四个点的位置
M_t = cv2.getPerspectiveTransform(pt_o, pt_d)
car_plate = cv2.warpPerspective(tmp_plate, M_t, (922,624)) # 最后一参数是输出dst的尺寸。可以和原来图片尺寸不一致。按需求来确定
plt.imshow(car_plate)
```

```python
tmp_car=car
for i in range(624):
    for j in range(922):
        if car_plate[i][j].all()!=False:
            tmp_car[i][j]=car_plate[i][j]
```
