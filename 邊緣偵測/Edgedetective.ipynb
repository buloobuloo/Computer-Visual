```python
from PIL import Image
import keras
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from skimage import exposure
```

```python
x = Image.open('IMG_6900.jpeg')
x=np.asarray(x)
```

```python
kernelx = np.array([[1,0,-1],[2,0,-2],[1,0,-1]])
kernely = np.array([[1,2,1],[0,0,0],[-1,-2,-1]])
```

```python
image = cv2.filter2D(x,-1,kernelx)
plt.imshow(image, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```

```python
image = cv2.filter2D(x,-1,kernely)
plt.imshow(image, cmap=plt.cm.gray)
plt.axis('off')
plt.show()
```
